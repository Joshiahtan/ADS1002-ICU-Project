```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("Preprocessed ICU data.csv")
df.head()
```

```python
df_x  = df.filter(regex="x$", axis = 1)


df_x
```

```python
keep_vars = ["GCS", "HR", "Platelets", "Urine", "WBC", "BUN", "Temp", "Glucose", "PaO2", "PaCO2","SysABP","HCO3","Weight", "Lactate"]

keep_vars_lower = [v.lower() for v in keep_vars]

filtered_cols = [
    col for col in df.columns
    if any(k in col.lower() for k in keep_vars_lower)
]

df_filtered = df[filtered_cols]  # keep target too

df_filtered
```

```python
filtered_cols_x = [
    col for col in df_x.columns
    if any(k in col.lower() for k in keep_vars_lower) 
]


df_filtered_x = df_x[filtered_cols_x]  # keep target too

df_filtered_x
```

```python
df_mean = pd.DataFrame(index=df_filtered.index)  # start new df

for col in df_filtered.columns:
    if col.startswith("Mean_") and col.endswith(".x"):
        base_name = col[:-2]  # remove ".x"
        col_x = base_name + ".x"
        col_y = base_name + ".y"

        if col_y in df_filtered.columns:
            # Mean across x and y
            df_mean[base_name + "_mean_xy"] = (df_filtered[col_x] + df_filtered[col_y]) / 2.0

            # Difference (y - x)
            df_mean[base_name + "_diff"] = df_filtered[col_y] - df_filtered[col_x]



df_mean.shape
```

```python
df_combined = pd.concat([df_filtered, df_mean], axis=1)

df_combined
```
