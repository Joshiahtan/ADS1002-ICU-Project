```python
from matplotlib.ticker import MaxNLocator

def many_box_plots(the_dataframe):
    numeric_columns = the_dataframe.select_dtypes(include='number').columns
    num_cols = len(numeric_columns)
    
    ncols = 3
    nrows = int(np.ceil(num_cols/ncols))
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize = (12,6*nrows))
    
    axes = axes.flatten() # changes to 1 D array
    
    
    for i in range(num_cols):
        col = numeric_columns[i]
        ax = axes[i]
        ax.boxplot(df[col].dropna())
        ax.set_ylabel(col)
        ax.set_xticks([])
        ax.yaxis.set_major_locator(MaxNLocator(nbins=10))

    
    for j in range(num_cols, len(axes)):
        axes[j].set_visible(False)
    plt.tight_layout()
    plt.show()
```

```python
many_box_plots(mean1_1)
```
