```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("Preprocessed ICU data.csv")
df.head()
```

```python
df_x  = df.filter(regex="x$", axis = 1)


df_x
```

```python
keep_vars = ["GCS", "HR", "Platelets", "Urine", "WBC", "BUN", "Temp", "Glucose", "PaO2", "PaCO2","SysABP","HCO3","Weight", "Lactate"]

keep_vars_lower = [v.lower() for v in keep_vars]

filtered_cols = [
    col for col in df.columns
    if any(k in col.lower() for k in keep_vars_lower)
]

df_filtered = df[filtered_cols]  # keep target too

df_filtered
```

```python
filtered_cols_x = [
    col for col in df_x.columns
    if any(k in col.lower() for k in keep_vars_lower) 
]


df_filtered_x = df_x[filtered_cols_x]  # keep target too

df_filtered_x
```

```python
df_mean = pd.DataFrame(index=df_filtered.index)  # start new df

for col in df_filtered.columns:
    if col.startswith("Mean_") and col.endswith(".x"):
        base_name = col[:-2]  # remove ".x"
        col_x = base_name + ".x"
        col_y = base_name + ".y"

        if col_y in df_filtered.columns:
            # Mean across x and y
            df_mean[base_name + "_mean_xy"] = (df_filtered[col_x] + df_filtered[col_y]) / 2.0

            # Difference (y - x)
            df_mean[base_name + "_diff"] = df_filtered[col_y] - df_filtered[col_x]



df_mean.shape
```

```python
df_combined = pd.concat([df_filtered, df_mean], axis=1)

df_combined
```

```python
df_combined.info
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
X = df_combined[[
    'Mean_GCS_mean_xy',
    'Max_GCS.y',
    'Mean_GCS.y',
    'Min_GCS.y',
    'Mean_NISysABP.y',
    'Mean_GCS_diff',
    'Mean_GCS.x',
    'Min_HR.x',
    'Mean_PaO2_mean_xy',
    'Mean_PaO2.x',
    'Mean_PaO2.y',
    'Mean_PaO2_diff',
    'Mean_HR_mean_xy',
    'Min_NISysABP.y',
    'Mean_Temp.y',
    'Mean_Temp_diff',
    'Max_PaO2.y',
    'Mean_HR.x',
    'Max_NISysABP.y',
    'Min_Temp.y'
]]
y = df["In.hospital_death"]
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.8, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

```python
X_train.dtypes
```

```python
X_test.head()
```

```python
rfc = RandomForestClassifier(n_estimators = 200, random_state = 42, oob_score = True)
```

```python
rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)

print(format(accuracy_score(y_test, y_pred)))
```

```python
rfc.oob_score_
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```

```python
importances = rfc.feature_importances_
feat_importances= pd.Series(importances, index = X.columns)
feat_importances = feat_importances.sort_values(ascending = False)
```
