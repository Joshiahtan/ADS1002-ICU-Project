```python
# import relevant libraries and modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None) 
```

```python
df = pd.read_csv("Preprocessed ICU data.csv")
df = df[~np.isclose(df["Gender"].round(8), 0.56192144,atol=1e-8)]
gen = df.iloc[:,:7]
mean1 = df.iloc[:,7:44]
min1 = df.iloc[:,44:81]
max1 = df.iloc[:,81:118]
mean2 = df.iloc[:,118:155]
min2 = df.iloc[:,155:192]
max2 = df.iloc[:,192:229]
outcome = df.iloc[:,229:]
```

```python
def do_rfe(X, y, n_features_to_select):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
    model = RandomForestClassifier(random_state=42)
    rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)
    rfe.fit(X_train,y_train)
    ranking=rfe.ranking_
    selected_features = X.columns[rfe.support_]
    X_train_selected = rfe.transform(X_train)
    X_test_selected = rfe.transform(X_test)
    model.fit(X_train_selected,y_train)
    y_pred=model.predict(X_test_selected)
    accuracy=accuracy_score(y_test,y_pred)
    importance = model.feature_importances_

    results = pd.DataFrame({
        'Feature': X.columns,
        'Ranking': ranking,
        'Importance': importance})
        
    # print("Selected Features: ", selected_features)
    # print(X.columns)
    # print(ranking)
    # print("Model accuracy with Selected Features:", accuracy)

    return results
    
    
```

```python
def do_rfe(X, y, n_features_to_select):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
    model = RandomForestClassifier(random_state=42)
    rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)
    rfe.fit(X_train,y_train)
    ranking=rfe.ranking_
    selected_features = X.columns[rfe.support_]
    X_train_selected = rfe.transform(X_train)
    X_test_selected = rfe.transform(X_test)
    model.fit(X_train_selected,y_train)
    y_pred=model.predict(X_test_selected)
    accuracy=accuracy_score(y_test,y_pred)
    # importance = model.feature_importances_

    results = pd.DataFrame({
        'Feature': X.columns,
        'Ranking': ranking})
        
    # print("Selected Features: ", selected_features)
    # print(X.columns)
    # print(ranking)
    # print("Model accuracy with Selected Features:", accuracy)

    return results
    
    
```

```python
datasets = [gen,mean1,min1,max1,mean2,min2,max2,outcome]
new_df = pd.DataFrame(index=df.index)
for dataset in datasets:
    keep_vars = ["GCS", "HR", "Platelets", "Urine", "WBC", "BUN", "Temp", "Glucose", "PaO2", "PaCO2","SysABP","HCO3","Weight", "Lactate"]
    base_name = f"{dataset}new"
    keep_vars_lower = [v.lower() for v in keep_vars]
    
    filtered_cols = [
        col for col in df.columns
        if any(k in col.lower() for k in keep_vars_lower)
    ]
    
    new_df = pd.concat([new_df, df[filtered_cols]], axis=1)



```

```python
new_df.columns
```

```python
new_df = new_df.loc[:, ~new_df.columns.duplicated()]
```

```python
df_meandif = pd.DataFrame(index=new_df.index)  # start new df

for col in new_df.columns:
    if col.startswith("Mean_") and col.endswith(".x"):
        base_name = col[:-2]  # remove ".x"
        col_x = base_name + ".x"
        col_y = base_name + ".y"

        if col_y in new_df.columns:
            # Mean across x and y
            df_meandif[base_name + "_mean_xy"] = (new_df[col_x] + new_df[col_y]) / 2.0

            # Difference (y - x)
            df_meandif[base_name + "_diff"] = new_df[col_y] - new_df[col_x]

```
