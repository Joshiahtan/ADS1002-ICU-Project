```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import confusion_matrix, precision_score, recall_score

df = pd.read_csv("Preprocessed ICU data.csv")

y = df["In.hospital_death"]
X = df.drop(columns=["In.hospital_death"])

def is_icu_type_234(col):
    c = col.strip().lower().replace(" ", "")
    return (c.startswith("icutype") or c.startswith("icu_type")) and c[-1:] in {"2", "3", "4"}

cols_to_drop = [c for c in X.columns if is_icu_type_234(c)]
if cols_to_drop:
    X = X.drop(columns=cols_to_drop, errors="ignore")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

model = LogisticRegression(max_iter=1000, solver="liblinear")

n_features_to_keep = max(1, int(X.shape[1] * 0.25))
selector = RFE(estimator=model, n_features_to_select=n_features_to_keep)
selector.fit(X_train, y_train)

y_pred = selector.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec = recall_score(y_test, y_pred, zero_division=0)

selected_mask = selector.support_
selected_features = X.columns[selected_mask]

model_selected = LogisticRegression(max_iter=1000, solver="liblinear")
model_selected.fit(X_train[selected_features], y_train)

coefs = np.abs(model_selected.coef_).ravel()
importance_df = pd.DataFrame({
    "feature": selected_features,
    "importance": coefs
}).sort_values("importance", ascending=False).reset_index(drop=True)

top_k = min(10, len(importance_df))
final_top10 = importance_df.head(top_k).copy()

print("\nFinal Top 10 Features")
print(final_top10.to_string(index=False))

all_rankings = []
for seed in range(10):
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=seed, stratify=y)
    sel = RFE(estimator=LogisticRegression(max_iter=1000, solver="liblinear"),
              n_features_to_select=n_features_to_keep)
    sel.fit(Xtr, ytr)
    ranking = pd.DataFrame({
        "feature": X.columns,
        "ranking": sel.ranking_,
        "random_state": seed
    })
    all_rankings.append(ranking)

rankings_df = pd.concat(all_rankings, ignore_index=True)
stability = (rankings_df.groupby("feature")["ranking"]
             .apply(lambda x: (x == 1).sum())
             .sort_values(ascending=False))
try:
    from IPython.display import display
    display(final_top10)
except:
    pass
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import confusion_matrix, precision_score, recall_score

df = pd.read_csv("Preprocessed ICU data.csv")

y = df["In.hospital_death"]
X = df.drop(columns=["In.hospital_death"])

def drop_icu_234(col):
    s = col.strip().lower().replace(" ", "")
    return (s.startswith("icutype") or s.startswith("icu_type")) and s[-1:] in {"2", "3", "4"}

cols = [c for c in X.columns if drop_icu_234(c)]
if cols:
    X = X.drop(columns=cols, errors="ignore")

X = X.select_dtypes(include=[np.number])
if X.isna().any().any():
    X = X.fillna(X.median(numeric_only=True))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

clf = LogisticRegression(max_iter=1000, solver="liblinear")
k = max(1, int(X.shape[1] * 0.25))
rfe = RFE(estimator=clf, n_features_to_select=k, step=1)
rfe.fit(X_train, y_train)

y_pred = rfe.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec = recall_score(y_test, y_pred, zero_division=0)

print("Confusion Matrix:\n", cm)
print(f"Precision: {prec:.3f}")
print(f"Recall:    {rec:.3f}")

sel_feats = X.columns[rfe.support_]
clf_sel = LogisticRegression(max_iter=1000, solver="liblinear")
clf_sel.fit(X_train[sel_feats], y_train)

imp = np.abs(clf_sel.coef_).ravel()
imp_df = (
    pd.DataFrame({"Feature": sel_feats, "Importance": imp})
    .sort_values("Importance", ascending=False, kind="mergesort")
    .reset_index(drop=True)
)

def top10_prefix_x(df_imp, prefix, all_cols):
    pat = rf"(?i)^{prefix}_"
    picked = df_imp[
        df_imp["Feature"].str.contains(pat, regex=True, na=False)
        & df_imp["Feature"].str.endswith(".x", na=False)
    ].copy()
    if len(picked) < 10:
        candidates = [c for c in all_cols if isinstance(c, str) and c.endswith(".x") and pd.Series([c]).str.contains(pat, regex=True, na=False).iloc[0]]
        missing = [c for c in candidates if c not in set(picked["Feature"])]
        if missing:
            picked = pd.concat([picked, pd.DataFrame({"Feature": missing, "Importance": 0.0})], ignore_index=True)
    return picked.head(10)

t_max  = top10_prefix_x(imp_df, "max",  X.columns)
t_mean = top10_prefix_x(imp_df, "mean", X.columns)
t_min  = top10_prefix_x(imp_df, "min",  X.columns)

def show(df_tbl, title):
    if df_tbl.empty:
        print(f"\n{title}\n(no matching .x features found)")
        return
    try:
        from IPython.display import display
        display(df_tbl.style.hide(axis="index").format({"Importance": "{:.3f}"}).set_caption(title))
    except:
        print(f"\n{title}")
        print(df_tbl.to_string(index=False))

show(t_max,  "Top 10 â€” MAX (.x) Features by Importance")
show(t_mean, "Top 10 â€” MEAN (.x) Features by Importance")
show(t_min,  "Top 10 â€” MIN (.x) Features by Importance")
```
